###**Preprocessing for olympic_results**

````python
df_results = pd.read_csv("/content/drive/MyDrive/DATA SCIENCE 23 24/2nd semester/Data managment/1st. Project/olympic_results.csv", sep = ",")
````

````python
#Dropped columns because of too many NULL values, unreadable and duplicate information
df_results = df_results.drop(columns =  ["rank_equal", "athlete_url", "country_code", "country_3_letter_code", "value_type", "value_unit"])
df_results
````

````python
#Delete rows from a DataFrame in Python where both athlete_full_name and athletes columns do not have values
df_results.dropna(subset=['athlete_full_name', 'athletes'], how='all', inplace=True)
````

````python
#Because the format fo the values was "[('Stefania CONSTANTINI', 'https://olympics.co..." and we were interested only in atheltes name we extracted them from "athletes"
#using Regex patterns
pattern = r"'([^',]+)'(?=, 'https)"
df_results['athletes']= df_results['athletes'].apply(lambda x: re.findall(pattern, x) if pd.notnull(x) and isinstance(x, str) else np.nan)
````

````python
df_results
````

````python
df_results.to_csv('olympic_results_new.csv', index = False)
````

###**Create index for the foreign key between medals and hosts, and between athletes and hosts**



```
hosts and medals
```



````python
df_medals = pd.read_csv("/content/drive/MyDrive/DATA SCIENCE 23 24/2nd semester/Data managment/1st. Project/new csv/olympic_medals_new.csv", sep = ";")
df_hosts = pd.read_csv("/content/drive/MyDrive/DATA SCIENCE 23 24/2nd semester/Data managment/1st. Project/new csv/olympic_hosts_new.csv", sep = ";")
df_athletes = pd.read_csv("/content/drive/MyDrive/DATA SCIENCE 23 24/2nd semester/Data managment/1st. Project/new csv/olympic_athletes_new.csv", sep = ";")
````

````python
#hosts - medals
merged_df = df_medals.merge(df_hosts, left_on='slug_game', right_on='game_slug', how='left')
merged_df.drop(["game_slug","game_end_date","game_start_date","game_location","game_name","game_season","game_year"], axis=1, inplace=True)
````



```
Athlete and medals
```



````python
#athletes - medals
merged_df = merged_df.merge(df_athletes, left_on='athlete_full_name', right_on='athlete_full_name', how='left')
merged_df.drop(["games_participations",	"first_game",	"athlete_year_birth", "Unnamed: 5"], axis=1, inplace=True)
````

````python
merged_df = merged_df.dropna(subset=['ID_athletes']) #delete rows that have ID_athletes = NaN

merged_df['ID_athletes'] = merged_df['ID_athletes'].astype(int)
````

````python
merged_df
````

````python
merged_df.to_csv('olympic_medals_new.csv', index = False)
````
